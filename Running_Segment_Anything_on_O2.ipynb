# Running Segment Anything (SAM) on O2

To use SAM on O2 with a Jupyter Notebook, start by setting up a virtual environment. While we recommend using a conda environment for this, the LMOD module is also an option. Keep in mind that Jupyter isn't pre-installed on O2, so you'll need to install it separately. To simplify this process, download the `.yaml` file from our GitHub page, which will automatically install Jupyter along with any other necessary packages. Follow these guidelines to prepare your environment for running SAM on O2.

## Installing SAM using the ```.yaml``` file


### Login to HMS O2

First you will need to login to your O2 account. You can find more information on how to login to O2 on the RC [resources website](https://harvardmed.atlassian.net/wiki/spaces/O2/pages/1601700123/How+to+login+to+O2).


### Connect to a compute node

```
srun --pty -p interactive -t 0-4:00 --mem 16g bash
```

We recommend you request for at least 16 GB memory and 4 hours of wall time

### Clone the GitHub repository


```
git clone https://github.com/rkarmaka/SAM-using-Python-for-Microscopy-Data.git

cd SAM-using-Python-for-Microscopy-Data
```

### Create a conda environment for SAM

```
module load miniconda3/23.1.0

conda env create -n sam -f sam-env.yaml
```

```-n``` specifies the name of the environment. If you want to have a specific name, you can mention it here. Otehrwise, the default name of the environment will be ```sam```. Please remember the environemnt name because in the future, you will need to access this environment to run SAM.
